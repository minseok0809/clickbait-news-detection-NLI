```python
import pandas as pd
import os
from glob import glob
```

```python
file_path = "./output/best"

all_data = pd.DataFrame()

for i, path in enumerate(glob(r"./best/*.csv")):
    df = pd.read_csv(path)
    all_data[i] = df.label
all_data
```

```python
hard_voting = all_data.mode(axis=1)
hard_voting.head()
```

```python
submission = pd.DataFrame({'index' : hard_voting.index, 'label': hard_voting[0]})
submission
```

```python
submission.to_csv(os.path.join(file_path, f"ensemble.csv"), index=False)
```
